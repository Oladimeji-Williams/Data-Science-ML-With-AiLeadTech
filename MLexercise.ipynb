```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression

scaler = MinMaxScaler()
lr = LinearRegression()

#load the dataset
USA_housing = pd.read_csv(r"C:\Users\Williams\Desktop\DataScienceWithAiLead\data\USA_housing.csv")
USA_housing
```

```python
#Question 1, number 3
sns.regplot(x="Avg. Area Income", y="Price",  data=USA_housing)
plt.title("Avg. Area Income vs Price")
plt.show()
sns.regplot(x="Avg. Area House Age", y="Price",  data=USA_housing)
plt.title("Avg. Area House Age vs Price")
plt.show()
sns.regplot(x="Avg. Area Number of Rooms", y="Price",  data=USA_housing)
plt.title("Avg. Area Number of Rooms vs Price")
plt.show()
sns.regplot(x="Avg. Area Number of Bedrooms", y="Price",  data=USA_housing)
plt.title("Avg. Area Number of Bedrooms vs Price")
plt.show()
sns.regplot(x="Area Population", y="Price",  data=USA_housing)
plt.title("Area Population vs Price")
plt.show()
```

```python
#Question 1, number 4
X = USA_housing[["Avg. Area Income", "Avg. Area House Age", "Avg. Area Number of Rooms", "Avg. Area Number of Bedrooms", "Area Population"]]
y = USA_housing["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=0)
```

```python
#Question 1, number 5
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Question 1, number 6
lr.fit(X_train_scaled, y_train)
lr.predict(X_test_scaled)
```

```python
#Question 1, number 7
print("The train accuracy is: {:.4f} ".format(lr.score(X_train_scaled, y_train)))
print("The test accuracy is: {:.4f} ".format(lr.score(X_test_scaled, y_test)))
```
