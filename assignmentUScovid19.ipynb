### Using the US Covid19 dataset, Select all the data where state is Washington, Virginia, and Kansas, and where county is snohomish and Orange

```python
import pandas as pd
url = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv"
covid = pd.read_csv(url)
covid.head()
```

```python
states = ["Washington", "Virginia", "Kansas"] #list of the desired states
counties = ["Snohomish", "Orange"] #list of desired counties
state_group = covid.groupby("state") #group by state and return dataframe in a variable
state_df = pd.DataFrame() #declare empty dataframe to hold dataframe containing only the desired states
result = pd.DataFrame() #declare empty dataframe to hold the final result

for i in state_group: #iterate over each observation in the state grouped dataframe
    for j in states: #iterate over each desired state
        state_df = state_df.append(state_group.get_group(j)) #append and save each observation containing each desired state

county_group = state_df.groupby("county") #group by county and return dataframe in a variable
county_df = pd.DataFrame() #declare empty dataframe to hold dataframe containing only the desired counties

for i in county_group: #iterate over each observation in the county grouped dataframe
    for j in counties: #iterate over each desired county
        county_df = county_df.append(county_group.get_group(j)) #append and save each observation containing each desired county
result = county_df.copy() #copy the result to a new variable
```
